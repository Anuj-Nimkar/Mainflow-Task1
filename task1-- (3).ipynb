```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'student-mat.csv'  
data = pd.read_csv(file_path, delimiter=';')

# Verify the columns and structure
print(data.columns)
print(data.head())
```

```python
# Check for missing values
print("Missing values in each column:\n", data.isnull().sum())

# Display data types
print("\nData types of each column:\n", data.dtypes)

# Check dataset size (rows, columns)
print("\nDataset shape (rows, columns):", data.shape)
```

```python
data = data.drop_duplicates()
print("\nDataset shape after removing duplicates:", data.shape)
```

```python
data.fillna(data.median(numeric_only=True), inplace=True)
print("\nMissing values handled.")
```

# Data Analysis Questions

```python
# 1. Average score in (G3)
avg_g3 = data['G3'].mean()
print(f"Average final grade (G3): {avg_g3}")
```

```python
# 2. Number of students scoring above 15 in final grade(G3)
high_scorers = data[data['G3'] > 15].shape[0]
print(f"Number of students scoring above 15 in G3: {high_scorers}")
```

```python
# 3. Correlation between study time and final grade
correlation = data['studytime'].corr(data['G3'])
print(f"Correlation between study time and final grade (G3): {correlation}")
```

```python
# 4.Gender with higher average final grade
avg_grades_by_gender = data.groupby('sex')['G3'].mean()
print("Average grades by gender:\n", avg_grades_by_gender)
```

# Data Visualization

```python
# 1. Histogram of final grades(G3)
plt.hist(data['G3'], bins=10, color='skyblue',edgecolor='black', alpha=0.7)
plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.grid(axis='y',alpha=0.7)
plt.show()
```

```python

# 2. Scatter plot between study time and final grade
sns.scatterplot(x='studytime', y='G3', data=data)
plt.title('Scatter Plot : Study Time vs. Final Grade')
plt.xlabel('Study Time (hours/week)')
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.show()
```

```python
#3. Bar chart comparing average scores of male and female students
avg_grades_by_gender.plot(kind='bar', color=['skyblue', 'orange'],edgecolor='black')
plt.title('Average Grades by Gender')
plt.ylabel('Average Grade (G3)')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.grid(axis='y',alpha=0.7)
plt.show()
```
